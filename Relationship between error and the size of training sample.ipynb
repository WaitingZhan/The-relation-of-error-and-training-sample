```python
def Generate_training_set(size_of_training_set):
    random_list = []
    for x in range(size_of_training_set):
        random_list.append(random.randrange(1,999))
    return random_list
```

```python
def Generate_Skewed_example(size_of_training_set,dense_min, dense_max, sparse_min, sparse_max):
    random_lower_example_list =[]
    for x in range(int(size_of_training_set*0.8)):
        random_lower_example_list.append(random.randrange(dense_min,dense_max))
    for x in range(int(size_of_training_set*0.2)):
        random_lower_example_list.append(random.randrange(sparse_min,sparse_max))
    return random_lower_example_list
```

```python
def inRange(aList):
    Range=[]
    for num in range(len(aList)):
        if 400<=aList[num]<=600:
            Range.append(aList[num])
    return Range
```

```python
def findMin(aList):
    Min = 999
    for num in range(len(aList)):
        if aList[num]< Min:
            Min = aList[num]
    return Min
```

```python
def findMax(aList):
    Max = 0
    for num in range(len(aList)):
        if aList[num]> Max:
            Max = aList[num]
    return Max
```

```python
def ErrorRate(hypothesis_min, hypothesis_max):
    
    if hypothesis_min == hypothesis_max:
        er = 0.199
    if hypothesis_min == 999 and hypothesis_max == 0:
        er = 0.2
    else:
        #print("The error Rate is ", ((hypothesis_min-400)+(600-hypothesis_max))/1000)
        er = ((hypothesis_min-400)+(600-hypothesis_max))/1000
    return er
```

```python
def Average_Error_Rate(Number_of_random_example, times):
    train_size = []
    train_set = []
    for i in range(times):
        train_size.append(Number_of_random_example)
    for i in range (len(train_size)):
        train_set.append(Generate_training_set(train_size[i])) 
    Error=[]
    for i in range (len(train_size)):
        Error.append(ErrorRate(findMin(inRange(train_set[i])),findMax(inRange(train_set[i]))))
    return sum(Error)/len(Error)
```
